Question -2 What is a scatter plot?

A graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
plt.xlabel ("x- Axis")
plt.ylabel ("y- Axis")
plt.title ("Title Graph")
```

```python
x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([0, 100, 200, 300, 400, 500])
y1 = np.array([50, 20, 40, 20, 60, 70])
y2 = np.array([10, 20, 30, 40, 50, 60])
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig = plt.figure()
gs = fig.add_gridspec(3, hspace=0)
axs = gs.subplots(sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x, y)
axs[1].plot(x, y1, 'o')
axs[2].plot(x, y2, '+')
axs[3].plot(x, y3, '+')

# Hide x labels and tick labels for all but bottom plot.
for ax in axs:
    ax.label_outer()
```

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
fig, ax = plt.subplots()
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
ax.bar(company, profit, label=bar_labels, color=bar_colors)
plt.show()
```

What is a Box Plot?
A Box Plot is the visual representation of the statistical five number summary of a given data set.

A Five Number Summary includes:

Minimum,
First Quartile,
Median (Second Quartile),
Third Quartile,
Maximum,

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]

# Create the box plot
plt.boxplot(data)

# Add labels to the x-axis ticks
plt.xticks([1, 2], ['Box 1', 'Box 2'])

# Add a title to the plot
plt.title('Box Plot')

# Display the plot
plt.show()
```
